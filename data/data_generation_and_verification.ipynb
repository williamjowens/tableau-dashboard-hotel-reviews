```python
# Install necessary packages
!pip install faker textblob pycountry

import pandas as pd
from faker import Faker
import random
from datetime import datetime, timedelta
from textblob import TextBlob
import pycountry
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chisquare

# Initialize Faker
fake = Faker()

# Define categorical fields
TRAVELLER_TYPES = ['Solo', 'Couple', 'Family', 'Business', 'Group']
ROOM_TYPES = ['Single', 'Double', 'Suite', 'Deluxe', 'Studio', 'Family Room']
STAY_TYPES = ['Leisure', 'Business', 'Extended', 'Short-term', 'Conference', 'Holiday']
RECOMMEND_OPTIONS = ['yes', 'no']
VERIFIED_OPTIONS = ['Verified', 'Not Verified']

# Define sentiments and corresponding ratings
SENTIMENTS = {
    'positive': {
        'content': [
            "The service was excellent and the staff were very friendly.",
            "Had a wonderful stay with comfortable rooms and great amenities.",
            "The location was perfect, close to all major attractions.",
            "Clean and well-maintained facilities. Highly recommended!",
            "Exceptional experience with top-notch customer service.",
            "Loved the spacious room and the stunning view from the balcony.",
            "The breakfast buffet had a wide variety of delicious options.",
            "Quick check-in process and the concierge was very helpful.",
            "The hotel exceeded my expectations in every way.",
            "Relaxing atmosphere and beautifully decorated rooms."
        ],
        'rating_range': (8, 10)
    },
    'neutral': {
        'content': [
            "It was an average stay with nothing exceptional.",
            "Rooms were okay, but the amenities were lacking.",
            "Service was acceptable, but could be improved.",
            "Location was decent, not too close or too far from attractions.",
            "Overall, a standard hotel experience.",
            "The room was clean but quite small.",
            "Had a satisfactory stay, nothing memorable.",
            "The hotel was alright, met my basic needs.",
            "Comfortable enough, but nothing stood out.",
            "The stay was fine, though some facilities were outdated."
        ],
        'rating_range': (5, 7)
    },
    'negative': {
        'content': [
            "The service was mediocre at best.",
            "Had a disappointing stay with unclean rooms and rude staff.",
            "The location was inconvenient and far from attractions.",
            "Poor maintenance and lack of amenities. Would not recommend.",
            "Terrible experience with subpar customer service.",
            "The room was dirty and not properly maintained.",
            "Had issues with the booking process and customer support.",
            "The breakfast was limited and not up to the mark.",
            "Noisy environment made it hard to sleep.",
            "The hotel did not meet my expectations in any aspect."
        ],
        'rating_range': (1, 4)
    }
}

# Define specific rating categories out of 5
SPECIFIC_RATINGS = ['cleanliness', 'staff_service', 'facilities',
                    'location', 'value_for_money', 'amenities']

# Define a list of 60 strategic countries for a major hotel chain
# Each country is assigned a weight based on its desirability/popularity
# Higher weight means higher representation in the dataset
HOTEL_COUNTRIES = [
    # North America (15)
    ("United States", 25),
    ("Canada", 15),
    ("Mexico", 10),
    ("Brazil", 8),
    ("Argentina", 6),
    ("Chile", 5),
    ("Colombia", 4),
    ("Peru", 3),
    ("Venezuela", 2),
    ("Cuba", 2),
    ("Panama", 2),
    ("Costa Rica", 2),
    ("Jamaica", 2),
    ("Dominican Republic", 1),
    ("Puerto Rico", 1),

    # Europe (15)
    ("United Kingdom", 20),
    ("France", 18),
    ("Germany", 15),
    ("Italy", 15),
    ("Spain", 12),
    ("Netherlands", 8),
    ("Switzerland", 6),
    ("Sweden", 5),
    ("Norway", 4),
    ("Denmark", 3),
    ("Portugal", 2),
    ("Belgium", 2),
    ("Austria", 2),
    ("Greece", 1),
    ("Ireland", 1),

    # Asia (15)
    ("China", 20),
    ("Japan", 18),
    ("South Korea", 15),
    ("India", 12),
    ("Singapore", 10),
    ("United Arab Emirates", 8),
    ("Thailand", 6),
    ("Malaysia", 5),
    ("Indonesia", 4),
    ("Vietnam", 3),
    ("Philippines", 2),
    ("Israel", 2),
    ("Saudi Arabia", 2),
    ("Turkey", 1),
    ("Hong Kong", 1),

    # Oceania (5)
    ("Australia", 15),
    ("New Zealand", 10),
    ("Fiji", 5),
    ("Papua New Guinea", 3),
    ("Samoa", 2),

    # Africa (10)
    ("South Africa", 10),
    ("Egypt", 8),
    ("Morocco", 7),
    ("Kenya", 6),
    ("Nigeria", 5),
    ("Tanzania", 4),
    ("Uganda", 3),
    ("Ghana", 2),
    ("Algeria", 2),
    ("Ethiopia", 1)
]

# Create separate lists for countries and their corresponding weights
COUNTRIES, COUNTRY_WEIGHTS = zip(*HOTEL_COUNTRIES)

# Define stay type durations (min_days, max_days)
STAY_TYPE_DURATIONS = {
    'Short-term': (1, 3),
    'Business': (1, 5),
    'Leisure': (3, 7),
    'Extended': (7, 30),
    'Conference': (2, 5),
    'Holiday': (5, 14)
}

def generate_review_header(sentiment):
    """
    Generate a meaningful review header based on sentiment.
    """
    positive_headers = [
        "Exceptional Stay!",
        "Highly Recommend This Hotel",
        "Fantastic Experience",
        "Outstanding Service",
        "A Memorable Visit",
        "Loved Every Moment",
        "Top-Notch Amenities",
        "Perfect Location and Service",
        "Exceeded Expectations",
        "Absolutely Wonderful"
    ]

    neutral_headers = [
        "Average Experience",
        "Decent Stay",
        "Satisfactory Service",
        "Nothing Special",
        "Room for Improvement",
        "Standard Hotel Stay",
        "Okay Facilities",
        "Acceptable Location",
        "Mediocre Service",
        "Just Fine"
    ]

    negative_headers = [
        "Disappointing Stay",
        "Would Not Recommend",
        "Poor Service",
        "Terrible Experience",
        "Not Up to the Mark",
        "Very Unsatisfied",
        "Needs Significant Improvement",
        "Unpleasant Stay",
        "Lacking Amenities",
        "Subpar Facilities"
    ]

    if sentiment == 'positive':
        return random.choice(positive_headers)
    elif sentiment == 'neutral':
        return random.choice(neutral_headers)
    else:
        return random.choice(negative_headers)

def generate_review_content(sentiment):
    """
    Generate review content based on sentiment.
    """
    return random.choice(SENTIMENTS[sentiment]['content'])

def determine_overall_rating(sentiment):
    """
    Determine overall rating out of 10 based on sentiment.
    """
    low, high = SENTIMENTS[sentiment]['rating_range']
    return random.randint(low, high)

def determine_specific_ratings(overall_rating):
    """
    Determine specific ratings out of 5 based on overall rating.
    """
    if overall_rating >= 8:
        return {category: random.randint(4,5) for category in SPECIFIC_RATINGS}
    elif overall_rating >=5:
        return {category: random.randint(3,4) for category in SPECIFIC_RATINGS}
    else:
        return {category: random.randint(1,2) for category in SPECIFIC_RATINGS}

def generate_stay_dates(stay_type, start_date, end_date):
    """
    Generate first and last stay dates based on stay_type.

    Parameters:
    - stay_type (str): Type of stay.
    - start_date (datetime): Earliest possible stay date.
    - end_date (datetime): Latest possible stay date.

    Returns:
    - tuple: (first_stay_date, last_stay_date)
    """
    # Convert end_date to date to match with fake.date_between, which returns date
    end_date = end_date.date()
    min_days, max_days = STAY_TYPE_DURATIONS[stay_type]

    # Calculate latest possible first_stay_date to ensure last_stay_date + 1 year <= end_date
    latest_possible_first_stay = end_date - timedelta(days=max_days + 365)
    if latest_possible_first_stay < start_date.date():
        latest_possible_first_stay = start_date.date() + timedelta(days=(end_date - start_date.date()).days // 2)

    first_stay_date = fake.date_between(start_date=start_date.date(), end_date=latest_possible_first_stay)
    stay_duration = random.randint(min_days, max_days)
    last_stay_date = first_stay_date + timedelta(days=stay_duration)

    # Ensure last_stay_date does not exceed end_date - 1 day
    if last_stay_date > end_date - timedelta(days=1):
        last_stay_date = end_date - timedelta(days=1)

    return first_stay_date.strftime('%m/%d/%Y'), last_stay_date.strftime('%m/%d/%Y')

def generate_review_date(last_stay_date, start_date, end_date):
    """
    Generate review_date based on last_stay_date.
    review_date should be after last_stay_date and within one year.

    Parameters:
    - last_stay_date (datetime): The last day of stay.
    - start_date (datetime): Earliest possible review date.
    - end_date (datetime): Latest possible review date.

    Returns:
    - str: review_date in mm/dd/yyyy format
    """
    min_review_date = last_stay_date + timedelta(days=1)
    max_review_date = last_stay_date + timedelta(days=365)

    # Adjust max_review_date if it exceeds the overall end_date
    if max_review_date > end_date:
        max_review_date = end_date

    if min_review_date > max_review_date:
        # If min_review_date exceeds max, set review_date to max_review_date
        review_date = max_review_date
    else:
        review_date = fake.date_between(start_date=min_review_date, end_date=max_review_date)

    return review_date.strftime('%m/%d/%Y')

def generate_reviews(num_entries=2000):
    """
    Generate hotel reviews dataset.

    Parameters:
    - num_entries (int): Number of reviews to generate.

    Returns:
    - pd.DataFrame: Generated dataset.
    """
    data = []
    sentiments = ['positive', 'neutral', 'negative']
    sentiment_weights = [0.5, 0.3, 0.2]  # More positive reviews

    # Define overall date range
    overall_start_date = datetime(2016, 1, 1)
    overall_end_date = datetime(2024, 10, 31)

    for _ in range(num_entries):
        sentiment = random.choices(sentiments, weights=sentiment_weights, k=1)[0]
        header = generate_review_header(sentiment)
        author = fake.name()

        # Random stay_type
        stay_type = random.choice(STAY_TYPES)

        # Generate first and last stay dates based on stay_type
        first_stay_str, last_stay_str = generate_stay_dates(stay_type, overall_start_date, overall_end_date)
        first_stay_date = datetime.strptime(first_stay_str, '%m/%d/%Y')
        last_stay_date = datetime.strptime(last_stay_str, '%m/%d/%Y')

        # Generate review_date based on last_stay_date
        review_date_str = generate_review_date(last_stay_date, overall_start_date, overall_end_date)

        # Format dates
        first_stay_date_formatted = first_stay_str
        last_stay_date_formatted = last_stay_str
        review_date = review_date_str

        place = random.choices(COUNTRIES, weights=COUNTRY_WEIGHTS, k=1)[0]
        content = generate_review_content(sentiment)
        room_type = random.choice(ROOM_TYPES)
        traveller_type = random.choice(TRAVELLER_TYPES)
        recommended = random.choice(RECOMMEND_OPTIONS)
        trip_verified = random.choice(VERIFIED_OPTIONS)
        rating = determine_overall_rating(sentiment)
        specific_ratings = determine_specific_ratings(rating)

        # Calculate 'composite_rating' as the average of specific ratings, rounded to one decimal
        composite_rating = round(
            (specific_ratings['cleanliness'] + specific_ratings['staff_service'] +
             specific_ratings['facilities'] + specific_ratings['location'] +
             specific_ratings['value_for_money'] + specific_ratings['amenities']) / 6, 1
        )

        data.append({
            'header': header,
            'author': author,
            'review_date': review_date,
            'place': place,
            'content': content,
            'room_type': room_type,
            'traveller_type': traveller_type,
            'stay_type': stay_type,
            'first_stay_date': first_stay_date_formatted,
            'last_stay_date': last_stay_date_formatted,
            'recommended': recommended,
            'trip_verified': trip_verified,
            'rating': rating,
            'cleanliness': specific_ratings['cleanliness'],
            'staff_service': specific_ratings['staff_service'],
            'facilities': specific_ratings['facilities'],
            'location': specific_ratings['location'],
            'value_for_money': specific_ratings['value_for_money'],
            'amenities': specific_ratings['amenities'],
            'composite_rating': composite_rating
        })

    df = pd.DataFrame(data)
    return df

def verify_reviews(df):
    """
    Verify the integrity and consistency of the generated dataset.

    Parameters:
    - df (pd.DataFrame): The dataset to verify.
    """
    print("=== Verification Report ===\n")

    # Check for missing values
    missing = df.isnull().sum()
    print("Missing values per column:")
    print(missing)

    # Check for duplicates
    duplicates = df.duplicated().sum()
    print(f"\nNumber of duplicate rows: {duplicates}")

    # Check date formats (mm/dd/yyyy)
    def check_date_format(date_str):
        try:
            datetime.strptime(date_str, '%m/%d/%Y')
            return True
        except:
            return False

    date_format_correct = (df['review_date'].apply(check_date_format).all() and
                           df['first_stay_date'].apply(check_date_format).all() and
                           df['last_stay_date'].apply(check_date_format).all())
    print(f"\nDate format (mm/dd/yyyy) correct: {date_format_correct}")

    # Check that first_stay_date <= last_stay_date <= review_date
    print("\nChecking date relationships...")
    df['first_stay_dt'] = pd.to_datetime(df['first_stay_date'], format='%m/%d/%Y')
    df['last_stay_dt'] = pd.to_datetime(df['last_stay_date'], format='%m/%d/%Y')
    df['review_dt'] = pd.to_datetime(df['review_date'], format='%m/%d/%Y')

    stay_order_correct = (df['first_stay_dt'] <= df['last_stay_dt']).all()
    review_after_stay_correct = (df['review_dt'] >= df['last_stay_dt']).all()
    review_within_one_year = (df['review_dt'] <= df['last_stay_dt'] + pd.Timedelta(days=365)).all()

    print(f"All first_stay_date <= last_stay_date: {stay_order_correct}")
    print(f"All review_date >= last_stay_date: {review_after_stay_correct}")
    print(f"All review_date within one year after last_stay_date: {review_within_one_year}")

    # Remove temporary datetime columns
    df.drop(['first_stay_dt', 'last_stay_dt', 'review_dt'], axis=1, inplace=True)

    # Check overall rating is between 1 and 10
    overall_rating_correct = df['rating'].between(1,10).all()
    print(f"\nOverall rating between 1-10: {overall_rating_correct}")

    # Check specific ratings are between 1 and 5
    specific_rating_correct = True
    for category in SPECIFIC_RATINGS:
        if not df[category].between(1,5).all():
            specific_rating_correct = False
            print(f"{category} ratings out of range.")

    if specific_rating_correct:
        print("All specific ratings are between 1-5.")
    else:
        print("There are specific ratings out of the expected range (1-5).")

    # Check consistency between sentiment and ratings
    # Using TextBlob to determine sentiment from 'content' and compare with 'rating'
    print("\nChecking consistency between review sentiment and overall rating...")
    sentiments_detected = df['content'].apply(lambda x: TextBlob(x).sentiment.polarity)

    # Define thresholds for sentiment
    # Polarity > 0.1: positive, < -0.1: negative, else neutral
    df['sentiment_polarity'] = sentiments_detected
    df['sentiment'] = df['sentiment_polarity'].apply(
        lambda x: 'positive' if x > 0.1 else ('negative' if x < -0.1 else 'neutral')
    )

    # Calculate average ratings per detected sentiment
    avg_ratings = df.groupby('sentiment')['rating'].mean()
    print("\nAverage Ratings by Detected Sentiment:")
    print(avg_ratings)

    # Remove the temporary columns
    df.drop(['sentiment_polarity', 'sentiment'], axis=1, inplace=True)

    # Check 'composite_rating' consistency
    # 'composite_rating' should be the average of specific ratings
    calculated_composite = df[['cleanliness', 'staff_service', 'facilities', 'location',
                               'value_for_money', 'amenities']].mean(axis=1).round(1)
    composite_consistent = (df['composite_rating'] == calculated_composite).all()
    print(f"\nComposite rating consistent with specific ratings: {composite_consistent}")

    # Verify Country Weights
    print("\nVerifying Country Weights:")

    # Calculate expected distribution based on weights
    total_weight = sum(COUNTRY_WEIGHTS)
    expected_probs = [weight / total_weight for weight in COUNTRY_WEIGHTS]

    # Get observed counts
    observed_counts = df['place'].value_counts().sort_index()

    # Ensure all countries are present in observed_counts
    observed_counts = observed_counts.reindex(COUNTRIES, fill_value=0)

    # Calculate expected counts
    expected_counts = [prob * len(df) for prob in expected_probs]

    # Create a DataFrame for comparison
    comparison_df = pd.DataFrame({
        'Country': COUNTRIES,
        'Weight': COUNTRY_WEIGHTS,
        'Expected_Count': expected_counts,
        'Observed_Count': observed_counts.values,
        'Expected_Proportion': expected_probs,
        'Observed_Proportion': observed_counts.values / len(df)
    })

    display(comparison_df.head(10))  # Display first 10

    # Plotting the distribution
    print("\nPlotting Expected vs Observed Country Distribution...")
    plt.figure(figsize=(14, 8))
    sns.barplot(x='Country', y='Observed_Proportion', data=comparison_df)
    plt.xticks(rotation=90)
    plt.title('Observed Country Distribution in "place" Column')
    plt.xlabel('Country')
    plt.ylabel('Observed Proportion')
    plt.tight_layout()
    plt.show()

    # Statistical Test - Chi-Squared Goodness-of-Fit
    print("\nPerforming Chi-Squared Goodness-of-Fit Test...")
    chi_stat, p_val = chisquare(f_obs=observed_counts, f_exp=expected_counts)
    print(f"Chi-Squared Statistic: {chi_stat}")
    print(f"P-Value: {p_val}")
    if p_val > 0.05:
        print("Result: Fail to reject the null hypothesis. Observed distribution matches expected distribution.")
    else:
        print("Result: Reject the null hypothesis. Observed distribution does not match expected distribution.")

    # Display sample data
    print("\nSample data (first 20 rows):")
    display(df.head(20))

    print("\n=== End of Verification ===")

def main():
    # Generate dataset with 2,000 entries
    reviews_df = generate_reviews(num_entries=2000)

    # Verify dataset
    verify_reviews(reviews_df)

    # Save to CSV
    reviews_df.to_csv('hotel_reviews.csv', index=False)
    print("\nDataset saved to 'hotel_reviews.csv'")

if __name__ == "__main__":
    main()
```
